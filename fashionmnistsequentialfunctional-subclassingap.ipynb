```python
import numpy as np
import tensorflow as tf
import keras
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Dense,Flatten
```

```python
(x_train,y_train),(x_test,y_test) = fashion_mnist.load_data()

print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
class_names = [
    'T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
    'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'
]

print(class_names[y_train[0]])

x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# Build a sequential API for building a model
model = Sequential()
model.add(Flatten(input_shape = (28,28)))
model.add(Dense(128,activation = "relu"))
model.add(Dense(128,activation = "softmax"))

model.compile(optimizer="adam",loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
model.summary()
```

```python
model.fit(x_train,y_train,epochs = 10,batch_size = 32)
```

```python
# Scaling is important for gradient descent. You get better results
x_test = x_test / 255.0 
x_test = np.reshape(x_test, (x_test.shape[0], 28, 28))
prediction = model.predict(x_test)
predicted_labels = np.argmax(prediction,axis = 1)
print("Predicted labels:", predicted_labels)
```

```python
input1 = keras.layers.Input(shape = (28,28))
flatten_layer = Flatten()(input1)
hidden1 = Dense(128,activation="relu")(flatten_layer)
hidden2 = Dense(64,activation="relu")(hidden1)
output = Dense(10,activation="softmax")(hidden2)

model = keras.Model(inputs = input1 , outputs = output)
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
model.summary()
```

```python
model.fit(x_train,y_train,epochs = 10,batch_size =32)
```

```python
class SubclassingAPI(keras.Model):
    def __init__(self):
        super(SubclassingAPI,self).__init__()
        self.flatten = Flatten()(self.input)
        self.hidden1 = Dense(128,activation="relu")(self.flatten)
        self.hidden2 = Dense(64,activation="relu")(self.hidden1)
        self.output = Dense(10,activation="softmax")(self.hidden2)
        
    def call(self,inputs):
        x = self.flatten(inputs)
        x = self.hidden1(x)
        x = self.hidden2(x)
        outputs = self.output(x)
        return outputs
    
model = SubclassingAPI()

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()
```
